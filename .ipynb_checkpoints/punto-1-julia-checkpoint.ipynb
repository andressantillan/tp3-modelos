```julia
using Pkg
Pkg.add("Distributions")
Pkg.add("Plots")
```

```julia
# Se utiliza el paquete Distributions para tener las funciones de distribucion
using Distributions, Statistics, Plots
```

```julia
# Utilizo la distribucion uniforme para la generacion de numeros uniformes
d = Uniform(0, 100)
```

```julia
# El primer parametro de rand es la funcion (en este caso la uniforme)
# Solo genera un numero
n = rand(d)
```

```julia
# Con el 2do parametro se inidca la cantidad de numeros aleatorios a generar
n = rand(d, 100)
```

```julia
# Calculo de la media
m = mean(n)
```

```julia
# Calculo del desvio estandar
desvio = std(n)
```

```julia
# Calculo de la varianza
varianza = var(n)
```

```julia
# La libreria para crear graficos con Julia es Plots
# Con plot, el primer argumento es la lista de valores para graficar 
plot(n; title="Valores aleatorios", ylabel="Valores")
```
